```python
import pandas as pd
```

```python
df = pd.read_csv('pico_web.csv', sep=';')
display(df)
display(df.head(4))
df.tail(4)
```

```python
df_copia = df.copy()
df_copia['Calories'] = df_copia['Calories'].fillna(0)
display(df_copia)
```

```python
df_copia['Date'] = df_copia['Date'].fillna('1900/01/01')
display(df_copia)
```

```python
df_copia['Date'] = pd.to_datetime(df_copia['Date'])
display(df_copia)
```
